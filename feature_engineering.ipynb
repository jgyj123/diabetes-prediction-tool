```python
import numpy as np
import pandas as pd

# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("diabetes_data.csv")
```

```python
df
```

```python
# get summary stats of age
df['Age'].describe()
```

```python
df_new = df.drop(['Genital thrush', 'delayed healing', 'partial paresis', 'Alopecia'], axis=1)
```

```python
df_new
```

```python
df_new['Age < 39'] = df_new['Age'] < 39  # Age is below 20,
df_new['Age 40-60'] = df_new['Age'].between(40, 59)  # Age is between 40 and 59, inclusive
df_new['Age >= 60'] = df_new['Age'] >= 60
```

```python
df_new.drop('Age', axis=1, inplace=True)
```

```python
le = LabelEncoder()
df_new['Gender'] = le.fit_transform(df_new['Gender'])
df_new['Polyuria'] = le.fit_transform(df_new['Polyuria'])
df_new['Polydipsia'] = le.fit_transform(df_new['Polydipsia'])
df_new['sudden weight loss'] = le.fit_transform(df_new['sudden weight loss'])
df_new['weakness'] = le.fit_transform(df_new['weakness'])
df_new['Polyphagia'] = le.fit_transform(df_new['Polyphagia'])
df_new['visual blurring'] = le.fit_transform(df_new['visual blurring'])
df_new['Itching'] = le.fit_transform(df_new['Itching'])
df_new['Irritability'] = le.fit_transform(df_new['Irritability'])
df_new['muscle stiffness'] = le.fit_transform(df_new['muscle stiffness'])
df_new['Obesity'] = le.fit_transform(df_new['Obesity'])
df_new['class'] = le.fit_transform(df_new['class'])
df_new['Age < 39'] = le.fit_transform(df_new['Age < 39'])
df_new['Age 40-60'] = le.fit_transform(df_new['Age 40-60'])
df_new['Age >= 60'] = le.fit_transform(df_new['Age >= 60'])
```

```python
df_new
```

```python
df_new.to_csv('diabetes_data_new_features.csv', index=False)
```
